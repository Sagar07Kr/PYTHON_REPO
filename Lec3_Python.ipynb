```python
#string functions
st=""
print(len(st))
# upper
s="hello"
s1=s.upper()
# lower
print(s1)
print(s1.lower())
# title
print(s1.title())
# capitalize()
S2="hello world"
print(S2.capitalize())

#strip function
# remove extra space from the front and end of the string
first = '   Anmol '
last = '   kuntal '
x= first.strip()
y= last.strip()
print(x+' '+y)
```

```python
a = " today "
b="  weather  "
c="is GooD"
a1=a.strip()
b1=b.strip()
s=a1+' '+b1+' '+c
x=s.lower()
y=x.title()
print(y)

# replace()
print(y.replace("Good","Bad"))

# find()
# find returns -1 if word does not present
print(y.find("Good"))
```
