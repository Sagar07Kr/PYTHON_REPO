```python
# Function in numpy .............
a= np.zeros(3)
print (a)
print(type(a[0]))

# this function gives a matrix of given dimension filled with 0
b= np.zeros((2,3))
print(b)

c=np.zeros((2,3,4))
print(c)

# by default the datatype of array's element is float we can change it by argument called dtype
d=np.zeros((2,3),dtype=int)
print(d)

# this function gives a matrix of given dimension filled with 1
e=np.ones(2)
print(e)

# this function gives a matrix of given dimension filled with random element
f=np.empty(4,dtype=int)
print(f)

# this function gives a matrix of given dimension filled with particulat number [size,element]
g=np.full(4,5,dtype=int)
print(g)
```

```python
# Question #
import numpy as np
a=[]
days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
for i in range(7):
  score=int(input(f"Score of {days[i]}:"))
  a.append(score)

c=np.array(a)

max_score=c.max()
print("Maximum score: ",max_score)

min_score=c.min()
print("Minimum score: ",min_score)

total_score=c.sum()
print("Total sum:",total_score)

score_mean=c.mean()
print("Mean:",score_mean)

std_score=c.std()
print("Standard Deviation:",std_score)




```

```python
# Homework ..........
import numpy as np

products = []
prices = []

n = int(input("Enter number of products: "))

for i in range(n):
    product = input(f"Enter name of product {i+1}: ")
    price = float(input(f"Enter price of {product}: "))

    products.append(product)
    prices.append(price)

p = np.array(prices)

print("\nProducts and Prices:")
for i in range(n):
    print(products[i], ":", prices[i])

print("\nHighest Price:", p.max())
print("Lowest Price:", p.min())
print("Total Price:", p.sum())
print("Average Price:", p.mean())
print("Standard Deviation:", p.std())
```
